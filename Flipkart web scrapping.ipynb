#    WEB SCRAPPING PROJECT ON FLIPKART

```python
# importing the library of web scrapping
from bs4 import BeautifulSoup as soup
import urllib.request as u
import pandas as pd
```

```python
# Setting and establishing the connnection yto the page url1
my_url1="https://www.flipkart.com/search?q=smartphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&p%5B%5D=facets.brand%255B%255D%3Drealme"
my_url1
```

```python
r=u.urlopen(my_url1)
r
```

```python
# reading html content of the page
ph=r.read()
ph
```

```python
r.close()
```

```python
#passing the html content into beautiful soup to fetch the pattern of the website
ps=soup(ph,"html.parser")
ps.prettify()
```

```python
# extracting product name
model1=[]
p_name1=ps.findAll('div',{'class':"_4rR01T"})
for i in p_name1:
    p=i.text
    model1.append(p)
print(model1)
len(model1)
```

```python
#extracting price
price1=[]
price_1=ps.findAll('div',{'class':"_30jeq3 _1_WHN1"})
for i in price_1:
    p=i.text
    price1.append(p)
print(price1)
len(price1)                          
```

```python
#extracting specification
specific1 = ps.findAll('div',{'class':"fMghEO"})
specific1
```

```python
# Extracting memory from specification
ram_rom1=[]
for i in specific1:
    p=i.findAll('li',{'class':'rgWa7D'})
    memory=p[0].text
    ram_rom1.append(memory)
print(ram_rom1)
len(ram_rom1)
```

```python
# Extracting display from specification
screen_size1=[]
for i in specific1:
    p=i.findAll('li',{'class':'rgWa7D'})
    display=p[1].text
    screen_size1.append(display)
print(screen_size1)
len(screen_size1)
```

```python
# Extracting camera from specification
camera1=[]
for i in specific1:
    p=i.findAll('li',{'class':'rgWa7D'})
    cam=p[2].text
    camera1.append(cam)
print(camera1)
len(camera1)
```

```python
# Extracting battery from specification
battery1=[]
for i in specific1:
    p=i.findAll('li',{'class':'rgWa7D'})
    bat=p[3].text
    battery1.append(bat)
print(battery1)
len(battery1)
```

```python
# Extracting processor from specification
processor1=[]
for i in specific1:
    p=i.findAll('li',{'class':'rgWa7D'})
    proc=p[4].text
    processor1.append(proc)
print(processor1)
len(processor1) 
```

```python
#extracting rating
rating1=[]
rating_1=ps.findAll('div',{'class':"_3LWZlK"})
for i in rating_1:
    p=i.text
    rating1.append(p)
print(rating1)
len(rating1) 
```

```python
data1=pd.DataFrame({"Model":model1,"Memory":ram_rom1,"Display":screen_size1,"Camera":camera1,"Battery":battery1,"Processor":processor1,"Price":price1,"Rating":rating1})
data1
```

```python
data1.to_csv('Dataset1.csv')
```
